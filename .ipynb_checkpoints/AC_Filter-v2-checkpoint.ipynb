#### Import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#### Create component classes for investigation and set constants

**Equations:**
- Capacitor Reactance: $\frac{1}{2 \pi freq C}$

```python
# Set values for mains frequency, create an array of resistance and capacitance values
frequency=50
resistor_values=[10e3,1.5e3]
capacitor_values=[1e-6, 0.1e-6]

class Resistor():
    '''Resistor component class'''

    def __init__(self, resistance):
        self.resistance = resistance

    def get_reactance(self):
        return self.resistance

    def __repr__(self):
        return f"Resistor of {self.resistance} Ohms"

class Capacitor():
    '''Capacitor component class'''

    def __init__(self, capacitance):
        self.capacitance = capacitance

    def get_reactance(self, frequency):
        return (1/ (np.pi * 2 * frequency * self.capacitance )) 

    def get_capacitance(self):
        return self.capacitance

    def __repr__(self):
        return f"Capacitor of {self.capacitance} Farad"

#create two instances of each class to use for the investigation
R1, R2, = Resistor(resistor_values[0]),Resistor(resistor_values[1])
C1, C2, = Capacitor(capacitor_values[0]),Capacitor(capacitor_values[1])
```

#### Create required functions

**Equations:**
- High-Pass filter phase : $ \phi = \arctan\left(\frac{1}{2 \pi freq R C}\right) $
 
- Low-Pass filter phase : $ \phi = -\arctan(2 \pi freq R C) $
  
- Cutoff frequency: $ f_c = \frac{1}{2 \pi R C} $

- High-Pass filter calculate_transfer function: $ H = \frac{2\pi fRC}{1+j2\pi fRC} \equiv\frac{V_{out}}{V_{in}}$

- Low-Pass filter calculate_transfer function: $ H = \frac{1}{1+j2\pi fRC} \equiv\frac{V_{out}}{V_{in}}$

```python
#Using equations, create functions which can be used for the pandas dataframe
def calculate_phaseshift(filter_type, freq, R, C):
    """
    Calculate phase shift for RC filters.
    Arguments:
        filter_type: 'low' (low-pass) or 'high' (high-pass)
        frequency: Input frequency (Hz)
        R: Resistance (Ohms)
        C: Capacitance (Farads)
    Returns:
        Phase shift in degrees
    """
    if filter_type == 'low':
        phi = np.arctan(2 * np.pi * freq * R * C)
        return -np.degrees(phi)
    elif filter_type == 'high':   
        phi = np.arctan(1 / (2 * np.pi * freq * R * C))
        return np.degrees(phi)

def cutoff_frequency(R,C):
    '''
    Calculate cutoff frequency for component combination
    Arguments:
        R: Resistance (Ohms)
        C: Capacitance (Farads)
    Returns:
        Cutoff frequency in Hertz'''
    fc = 1/(2 * np.pi * R * C)
    return (fc) 

def calculate_transfer(filter_type, freq, R, C):
    """
    Calculate transfer function for RC filters.
    Arguments:
        filter_type: 'low' (low-pass) or 'high' (high-pass)
        frequency: Input frequency (Hz)
        R: Resistance (Ohms)
        C: Capacitance (Farads)
    Returns:
        Transfer function (unitless)
    """
    if filter_type == 'low':
        H = 1 / (1 + 1j * 2 * np.pi * freq * R * C)
        return np.abs(H)
    elif filter_type == 'high': 
        H = (1j * 2 * np.pi * freq * R * C ) / (1 + 1j * 2 * np.pi * freq * R * C)
        return 20 * np.log10(np.abs(H))
```

#### Use functions to create arrays of data which will be used for pandas dataframe

**Equations:**
- Circuit Impedance: $Z=\sqrt{R^{2}+X_C^{2}}$

```python
#calculate circuit impedance to use in both pandas dataframes
Z=[(R1.get_reactance()**2+(C1.get_reactance(frequency))**2)**0.5/1000,(R2.get_reactance()**2+
        (C2.get_reactance(frequency))**2)**0.5/1000]

#calculate cutoff frequency for each combination
fc =[cutoff_frequency(R1.get_reactance(), C1.get_capacitance()),
     cutoff_frequency(R2.get_reactance(), C2.get_capacitance())]

#create arrays of values for the high-pass filter phase shift and voltage ratio
phase_high = [calculate_phaseshift('high', frequency, R1.get_reactance(), C1.get_capacitance()),
              calculate_phaseshift('high', frequency, R2.get_reactance(), C2.get_capacitance())]
transfer_high = [calculate_transfer('high', frequency, R1.get_reactance(), C1.get_capacitance()),
                 calculate_transfer('high', frequency, R2.get_reactance(), C2.get_capacitance())]

#create arrays of values for the low-pass filter phase shift and voltage ratio
phase_low = [calculate_phaseshift('low', frequency, R1.get_reactance(), C1.get_capacitance()),
              calculate_phaseshift('low', frequency, R2.get_reactance(), C2.get_capacitance())]
transfer_low = [calculate_transfer('low', frequency, R1.get_reactance(), C1.get_capacitance()),
              calculate_transfer('low', frequency, R2.get_reactance(), C2.get_capacitance())]
```

#### Use pandas to create and define data tables for filter plot visual

```python
#set data values for the high-pass filter table
hdata = {
    r'$R$ | $k\Omega$': [R1.get_reactance()/1000,R2.get_reactance()/1000],
    r'$X_C$ | $k\Omega$': [C1.get_reactance(frequency)/1000, C2.get_reactance(frequency)/1000],
    r'$Z$ | $k\Omega$': Z,
    r'$V_{out}/V_{in}$':transfer_high,
    r'Phase $\phi$ | $\degree$':phase_high,
    r'Cut-off $f_c$ | $kHz$':[fc[0]/1000, fc[1]/1000]
}

#create the high-pass filter dataframe
df_high = pd.DataFrame(hdata).transpose()
df_high=df_high.round(2) 
df_high.columns = ['Config 1', 'Config 2']

#set data values for the low-pass filter table
ldata = {
    r'$R$ | $\Omega$': [R1.get_reactance()/1000,R2.get_reactance()/1000],
    r'$X_C$ | $k\Omega$': [C1.get_reactance(frequency)/1000, C2.get_reactance(frequency)/1000],
    r'$Z$ | $k\Omega$': Z,
    r'$V_{out}/V_{in}$':transfer_low,
    r'Phase $\phi$ | $\degree$':phase_low,
    r'Cut-off $f_c$ | $kHz$':[fc[0]/1000, fc[1]/1000]
}
#create the low-pass filter dataframe 
df_low = pd.DataFrame(ldata).transpose()
df_low =df_low.round(2)
df_low.columns = ['Config 1', 'Config 2']
```

#### Display High-Pass filter data

Setup figures and create necessary variables for sine waves

```python
# Define necessary variables
frequency_overplot = np.logspace(0, 5, 500)
t = np.linspace(0, 0.1, 1000)

# Find closest frequency index
closest_index = np.argmin(np.abs(frequency_overplot - frequency))
closest_frequency = frequency_overplot[closest_index]

# Generate input sine wave
input = np.sin(2 * np.pi * closest_frequency * t)

# Compute output waves
output1 = df_high.iloc[3]['Config 1'] * np.sin(2 * np.pi * closest_frequency * t + 
                                               np.radians(df_high.iloc[4]['Config 1']))
output2 = df_high.iloc[3]['Config 2'] * np.sin(2 * np.pi * closest_frequency * t + 
                                               np.radians(df_high.iloc[4]['Config 2']))

# Setup figure
figh, axes_highpass = plt.subplots(4,2, figsize=(15,10))
(ax1_htable, ax2_htable),(ax1_mh, ax2_mh),(ax1_ph, ax2_ph),(ax1_tdh,ax2_tdh) = axes_highpass

figh.tight_layout()
figh.subplots_adjust(wspace=0.2, hspace=0.4) #adjust subplot locations
figh.suptitle('RC High-Pass Filter Response', fontsize=20, weight='bold')
plt.close()
```

**High-Pass filter tables**

```python
# Create configuration 1 table in the top left
ax1_htable.axis('off')
htable1 = ax1_htable.table(cellText=df_high.iloc[:,[0]].values, rowLabels=df_high.iloc[:,[0]].index, 
                          cellLoc='center', loc='center', bbox=[0.38, -0.1, 0.5, 0.8])
htable1.set_fontsize(14)
ax1_htable.set_title("Configuration 1")
#Create configuration 2 table in the top right
ax2_htable.axis('off')
htable2 = ax2_htable.table(cellText=df_high.iloc[:,[1]].values, rowLabels=df_high.iloc[:,[1]].index, 
                          cellLoc='center', loc='center',  bbox=[0.38, -0.1, 0.5, 0.8])
htable2.set_fontsize(14)
ax2_htable.set_title("Configuration 2")
```

**High-Pass filter configuration 1**

```python
#Magnitude response plot
ax1_mh.semilogx(frequency_overplot, calculate_transfer('high', frequency_overplot, R1.get_reactance(), C1.get_capacitance()), 'b', label='Gain')
ax1_mh.axvline(fc[0], color='r', linestyle="--", label=f"Cutoff = {df_high.iloc[5]['Config 1']:.2f} kHz")
ax1_mh.set(xlabel=r'frequency $Hz$', ylabel='Gain $dB$')
ax1_mh.legend()
ax1_mh.grid(True, which='both')

#Phase response plot
ax1_ph.semilogx(frequency_overplot, calculate_phaseshift('high', frequency_overplot, R1.get_reactance(), C1.get_capacitance()), label="Phase φ(freq)", color='g')
ax1_ph.axvline(fc[0], color='r', linestyle="--")
ax1_ph.set(xlabel=r'frequency $Hz$',ylabel=r'Phase $\phi \degree$')
ax1_ph.grid(True, which='both')
ax1_ph.legend()

#Time response plot
ax1_tdh.plot(t, input, 'b', label=r'Unfiltered $V_{in}$')
ax1_tdh.plot(t, output1, 'r', label=r'Filtered $V_{out}$')
ax1_tdh.set(xlabel='Time $s$', ylabel='Amplitude')
ax1_tdh.grid(True)
ax1_tdh.legend()
```

**High-Pass filter configuration 2**

```python
#Magnitude response plot
ax2_mh.semilogx(frequency_overplot, calculate_transfer('high', frequency_overplot, R2.get_reactance(), C2.get_capacitance()), 'b', label='Gain')
ax2_mh.axvline(fc[1], color='r', linestyle="--", label=f"Cutoff = {df_high.iloc[5]['Config 2']:.2f} kHz")
ax2_mh.set(xlabel=r'frequency $Hz$', ylabel='Gain $dB$')
ax2_mh.legend()
ax2_mh.grid(True, which='both')

#Phase response plot
ax2_ph.semilogx(frequency_overplot, calculate_phaseshift('high', frequency_overplot, R2.get_reactance(), C2.get_capacitance()), label="Phase φ(freq)", color='g')
ax2_ph.axvline(fc[1], color='r', linestyle="--")
ax2_ph.set(xlabel=r'frequency $Hz$',ylabel=r'Phase $\phi \degree$')
ax2_ph.grid(True, which='both')
ax2_ph.legend()

#Time response plot
ax2_tdh.plot(t, input, 'b', label=r'Unfiltered $V_{in}$')
ax2_tdh.plot(t, output2, 'r', label='Filtered $V_{out}$')
ax2_tdh.set(xlabel='Time $s$', ylabel='Amplitude')
ax2_tdh.grid(True)
ax2_tdh.legend()
figh
```
