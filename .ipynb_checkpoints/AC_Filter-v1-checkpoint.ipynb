#### Import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#### Create component classes for investigation and set constants

**Equations:**
- Capacitor Reactance: $\frac{1}{2 \pi f C}$

```python
# Set values for mains frequency and voltage, create an array of resistance and capacitance values
frequency=50
voltage=230
resistances=[100,1000]
capacitance=[10e-9, 100e-9]

class Resistor():
    '''Resistor component class'''

    def __init__(self, resistance):
        self.resistance = resistance

    def get_reactance(self):
        return self.resistance

    def __repr__(self):
        return f"Resistor of {self.resistance} Ohms"

class Capacitor():
    '''Capacitor component class'''

    def __init__(self, capacitance):
        self.capacitance = capacitance

    def get_reactance(self, frequency):
        return (1/ (np.pi * 2 * frequency * self.capacitance ))/1000 #return kOhm

    def get_capacitance(self):
        return self.capacitance

    def __repr__(self):
        return f"Capacitor of {self.capacitance} Farad"

#create three instances of each class to use for the investigation
R1, R2, = Resistor(resistances[0]),Resistor(resistances[1])
C1, C2, = Capacitor(capacitance[0]),Capacitor(capacitance[1])
```

#### Create required functions

**Equations:**
- High-Pass filter phase : $ \phi = \arctan\left(\frac{1}{2 \pi f R C}\right) $
 
- Low-Pass filter phase : $ \phi = -\arctan(2 \pi f R C) $
  
- Cutoff frequency: $ f_c = \frac{1}{2 \pi R C} $

- High-Pass filter transfer function: $ H = \frac{2\pi fRC}{1+j2\pi fRC} \equiv\frac{V_{out}}{V_{in}}$

- Low-Pass filter transfer function: $ H = \frac{1}{1+j2\pi fRC} \equiv\frac{V_{out}}{V_{in}}$

```python
#Using equations, create functions which can be used for the pandas dataframe
def phaseshift(filtertype, f, R, C):
    '''Calculate the phaseshift for a low-pass filter or high-pass filter'''
    if filtertype == 'low':
        phi = -np.arctan(2 * np.pi * f * R * C)
        return np.degrees(phi)
    elif filtertype == 'high':   
        phi = -np.arctan(1 / (2 * np.pi * f * R * C))
        return np.degrees(phi)

def cutoff_frequency(R,C):
    '''Calculate cutoff frequency for component combination'''
    fc = 1/(2 * np.pi * R * C)
    return (fc/1000) #return kHz

def transfer(filtertype, f, R, C):
    '''Calculate transfer function for plot, equivalent to voltage ratio'''
    if filtertype == 'low':
        H = 1 / (1 + 1j * 2 * np.pi * f * R * C)
        return np.abs(H)
    elif filtertype == 'high': 
        H = (1j * 2 * np.pi * f * R * C ) / (1 + 1j * 2 * np.pi * f * R * C)
        return np.abs(H) 
```

#### Use functions to create arrays of data which will be used for pandas dataframe

**Equations:**
- Circuit Impedance: $Z=\sqrt{R^{2}+X_C^{2}}$

```python
#calculate circuit impedance to use in both pandas dataframes
Z=[(R1.get_reactance()**2+(C1.get_reactance(frequency))**2)**0.5,(R2.get_reactance()**2+
        (C2.get_reactance(frequency))**2)**0.5]

#calculate cutoff frequency for each combination
fc =[cutoff_frequency(R1.get_reactance(), C1.get_capacitance()),
     cutoff_frequency(R2.get_reactance(), C2.get_capacitance())]

#create arrays of values for the high-pass filter phase shift and voltage ratio
phase_high = [phaseshift('high', frequency, R1.get_reactance(), C1.get_capacitance()),
              phaseshift('high', frequency, R2.get_reactance(), C2.get_capacitance())]
transfer_high = [transfer('high', frequency, R1.get_reactance(), C1.get_capacitance()),
                 transfer('high', frequency, R2.get_reactance(), C2.get_capacitance())]

#create arrays of values for the low-pass filter phase shift and voltage ratio
phase_low = [phaseshift('low', frequency, R1.get_reactance(), C1.get_capacitance()),
              phaseshift('low', frequency, R2.get_reactance(), C2.get_capacitance())]
transfer_low = [transfer('low', frequency, R1.get_reactance(), C1.get_capacitance()),
              transfer('low', frequency, R2.get_reactance(), C2.get_capacitance())]
```

#### Use pandas to create and define data tables for filter plot visual

```python
#set data values for the high-pass filter table
hdata = {
    r'$R$ | $\Omega$': [R1.get_reactance(),R2.get_reactance()],
    r'$X_C$ | $k\Omega$': [C1.get_reactance(frequency), C2.get_reactance(frequency)],
    r'$Z$ | $k\Omega$': Z ,
    r'$\frac{V_{out}}{V_{in}}$':transfer_high,
    r'Phase $\phi$ | $\degree$':phase_high,
    r'Cut-off $f_c$ | $kHz$':fc
}

#create the high-pass filter dataframe
df_high = pd.DataFrame(hdata).transpose()
df_high=df_high.round(2) 
df_high.columns = ['Config 1', 'Config 2']

#set data values for the low-pass filter table
ldata = {
    r'$R$ | $\Omega$': [R1.get_reactance(),R2.get_reactance()],
    r'$X_C$ | $k\Omega$': [C1.get_reactance(frequency), C2.get_reactance(frequency)],
    r'$Z$ | $k\Omega$': Z ,
    r'$\frac{V_{out}}{V_{in}}$':transfer_low,
    r'Phase $\phi$ | $\degree$':phase_low,
    r'Cut-off $f_c$ | $kHz$':fc
}
#create the low-pass filter dataframe 
df_low = pd.DataFrame(ldata).transpose()
df_low =df_low.round(2)
df_low.columns = ['Config 1', 'Config 2']
```

#### Setup layout for graphs and table

```python
#setup an array of subplots
figh, axes_highpass = plt.subplots(3,2, figsize=(20,12))
(_, __),(ax1_hplot,ax1_vhr),(ax2_hplot,ax2_vhr) = axes_highpass
figh.delaxes(_), figh.delaxes(__) #delete upper row of subplots

#setup matplotlib table that spans top row
ax_htable = figh.add_subplot(3,2,(1,2))
ax_htable.axis('off')
htable = ax_htable.table(cellText=df_high.values, rowLabels=df_high.index, colLabels=df_high.columns, cellLoc='center', loc='center')
htable.set_fontsize(12)
htable.scale(0.75,1.5)

#phase and v ratio plots
x = np.linspace(0, 1, 10000)
ax1_vhr.plot(x, np.sin(2 * np.pi * frequency * x),'k',label = f'Sine Wave')
ax1_vhr.plot(x, np.sin(2 * np.pi * frequency * (x+phase_high[0])),'--', label = f'Phase {float(phase_high[0]):.1f} degrees')
ax1_vhr.set(ylim=(-1,1), xlim=(0,0.05))
ax1_vhr.legend()

ax2_vhr.plot(x, np.sin(2 * np.pi * frequency * x),'k',label = f'Sine Wave')
ax2_vhr.plot(x, np.sin(2 * np.pi * frequency * (x+phase_high[1])),'--', label = f'Phase {float(phase_high[1]):.1f} degrees')
ax2_vhr.set(ylim=(-1,1), xlim=(0,0.05))
ax2_vhr.legend()

#figure methods to improve visual
figh.tight_layout()
figh.subplots_adjust(top=0.95) #stop table from clipping title
figh.suptitle('High-Pass Filter', fontsize=15, weight='bold')
```

```python
figl, axes_lowpass = plt.subplots(3,2, figsize=(20,12))
(_, __),(ax1_lplot,ax1_vlr),(ax2_lplot,ax2_vlr) = axes_lowpass
figl.delaxes(_), figl.delaxes(__)

ax_ltable = figl.add_subplot(3,2,(1,2))
ax_ltable.axis('off')
ltable= ax_ltable.table(cellText=df_low.values, rowLabels=df_low.index, colLabels=df_low.columns, cellLoc='center', loc='center')
ltable.scale(0.75,1.5)
ltable.set_fontsize(12)

#phase and v ratio plots
x = np.linspace(0, 1, 10000)
ax1_vlr.plot(x, np.sin(2 * np.pi * frequency * x),'k',label = f'Sine Wave')
ax1_vlr.plot(x,np.sin(2 * np.pi * frequency * (x+phase_low[0])), '--', label = f'Phase {float(phase_low[0]):.1f} degrees')
ax1_vlr.set(ylim=(-1,1), xlim=(0,0.05))
ax1_vlr.legend()

ax2_vlr.plot(x, np.sin(2 * np.pi * frequency * x),'k',label = f'Sine Wave')
ax2_vlr.plot(x, np.sin(2 * np.pi * frequency * (x+phase_low[1])),'--', label = f'Phase {float(phase_low[1]):.1f} degrees')
ax2_vlr.set(ylim=(-1,1), xlim=(0,0.05))
ax2_vlr.legend()

figl.tight_layout()
figl.subplots_adjust(top=0.95) #stop table from clipping title
figl.suptitle('Low-Pass Filter', fontsize=15, weight='bold')
```
